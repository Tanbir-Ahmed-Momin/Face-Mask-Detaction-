```
#importing the libraries
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import shutil, sys, dlib, glob
```

```
#setting the path to the directory containing the pics
path = '/content/drive/MyDrive/face/without_mask'
```

```


for f in glob.glob(os.path.join(path, "*.jpg")):
    shutil.copy2(f, path + 'temp.jpg')
    img = cv2.imread(path + 'temp.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    os.remove(path + 'temp.jpg')
```

```
#appending the pics to the training data list
training_data = []
for img in os.listdir(path):
    pic = cv2.imread(os.path.join(path,img))
    pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
    pic = cv2.resize(pic,(80,80))
    training_data.append([pic])
    #if(len(training_data)>=500)
   # break
```

```
#converting the list to numpy array and saving it to a file using #numpy.save
np.save(os.path.join(path,'featurs'),np.array(training_data))
```

```
#loading the saved file once again
saved = np.load(os.path.join(path,'featurs.npy'))
```
